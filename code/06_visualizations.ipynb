# Imports

```python
import matplotlib.pyplot as plt
import pandas as pd
import os
```

```python
%config InlineBackend.figure_format ='retina'
```

```python
df = pd.read_csv('datasets/train_clean.csv')
```

# Make a visualizations folder

```python
try:
    os.mkdir('visualizations')
except FileExistsError:
    pass
```

# Top 6 variables most highly correlated with `SalePrice`

```python
corrs = df.corr()[['SalePrice']].drop('SalePrice').sort_values(by='SalePrice',
                                                               ascending=False)[:6]
```

```python
x_vals = corrs.index[::-1]
y_vals = [corrs.loc[ix, 'SalePrice'] for ix in x_vals]
```

```python
plt.figure(figsize=(8, 8))
plt.barh(x_vals, y_vals)
plt.title('Top 6 variables most highly correlated with sale price', fontsize=20)
plt.xlabel('Strength of correlation', fontsize=15)
plt.savefig('visualizations/corrs.png');
```
